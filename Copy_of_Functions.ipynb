```python

# 24 Hariom Patidar
def count_vowels(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
    print(count)

count_vowels("Python")
```

```python

# 25 Hariom Patidar
def factorial(num):
    fact = 1
    for i in range(1, num + 1):
        fact *= i
    print(fact)

factorial(5)
```

```python
# 27 Hariom Patidar
def check_palindrome(text):
    if text == text[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrome")

check_palindrome("madam")
```
