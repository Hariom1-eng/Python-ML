Task1 write a program in python to calculate sides are given

```python
import math

# Input sides of triangle
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

# Calculate semi-perimeter
s = (a + b + c) / 2

# Calculate area using Heron's formula
area = math.sqrt(s * (s - a) * (s - b) * (s - c))

# Output result
print("Area of triangle =", area)
```

```python
# Input from user
ch = input("Enter a character: ")

# Convert to lowercase (so it works for both uppercase & lowercase)
ch = ch.lower()

# Check if it is a single alphabet
if len(ch) == 1 and ch.isalpha():

    if ch in 'aeiou':
        print("It is a vowel character")
    else:
        print("It is a consonant character")

else:
    print("Invalid input (please enter a single alphabet)")
```

write a program to check wether given charter s a digit or a  charter in lowecase or uppercase alphbet

```python
# Input from user
ch = input("Enter a character: ")

# Check if single character
if len(ch) == 1:
    if ch.isdigit():
        print("It is a digit")
    elif ch.islower():
        print("It is a lowercase alphabet")
    elif ch.isupper():
        print("It is an uppercase alphabet")
    else:
        print("It is a special character")
else:
    print("Please enter only one character")
```

Task 5 write a program in python to eval;ute  factorial


```python
n = int(input("Enter a number: "))

fact = 1
if n < 0:
    print("Factorial not defined for negative numbers")
else:
    for i in range(1, n + 1):
        fact = fact * i

    print("Factorial =", fact)
```

```python
max_num = None

while True:
    n = int(input("Enter number: "))

    if max_num is None or n > max_num:
        max_num = n

    choice = input("Do you want to continue? (y/n): ")
    if choice.lower() != 'y':
        break

print("Maximum number =", max_num)
```

```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    # Displaying info
    def display_info(self):
        print(f"Car: {self.brand} {self.model} ({self.year})")


# Creating object
car2 = Car("Tesla", "Model S", 2023)

# Calling method
car2.display_info()
```

```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    def display_info(self):
        print(f"Name: {self.name}, Roll: {self.roll}, Marks: {self.marks}")


# Creating list of student objects
students = []

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter name: ")
    roll = int(input("Enter roll number: "))
    marks = float(input("Enter marks: "))

    s = Student(name, roll, marks)
    students.append(s)


# Display all students
print("\nStudent Details:")
for s in students:
    s.display_info()
```
